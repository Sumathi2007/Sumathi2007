```python
# ðŸ“Š TASK 2 - Data Analysis
# Name: (Your Name)
# Date: 5 Nov

# Step 1ï¸âƒ£: Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2ï¸âƒ£: Create & save a sample CSV
data = {
    "Name": ["Riya", "Arjun", "Divya", "Karan", "Priya", "Vishal", "Meena"],
    "Department": ["CSE", "ECE", "CSE", "MECH", "CSE", "EEE", "CSE"],
    "Marks": [85, 67, 92, 55, 78, 60, 88],
    "Gender": ["F", "M", "F", "M", "F", "M", "F"]
}

df = pd.DataFrame(data)
df.to_csv("students.csv", index=False)  # Save CSV file

print("âœ… CSV file 'students.csv' created successfully!\n")

# Step 3ï¸âƒ£: Load the CSV file
data = pd.read_csv("students.csv")
print("ðŸ“‚ Loaded Data:")
print(data)

# Step 4ï¸âƒ£: Filtering data
print("\nðŸŽ¯ Students with marks > 70:")
print(data[data["Marks"] > 70])

print("\nðŸŽ¯ Students from CSE Department:")
print(data[data["Department"] == "CSE"])

# Step 5ï¸âƒ£: Visualization - Bar chart for marks
plt.figure(figsize=(8,5))
plt.bar(data["Name"], data["Marks"], color='lightblue')
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Students Marks Comparison")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

# Step 6ï¸âƒ£: Visualization - Department wise average marks
avg_marks = data.groupby("Department")["Marks"].mean()
plt.figure(figsize=(6,4))
avg_marks.plot(kind='bar', color='orange')
plt.title("Average Marks by Department")
plt.ylabel("Average Marks")
plt.show()

print("\nâœ… Task 2 completed successfully!")
```
